# Import libraries
- Seaborn automatically installs these libraries
    - numpy
    - scipy
    - pandas
    -matplotlib




```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris") #iris is a flower dataset
phool

# draw a line plot
sns.lineplot(x = "sepal_length", y="sepal_width", data = phool)
plt.show()
```

## Adding titles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris") #iris is a flower dataset
phool

# draw a line plot
sns.lineplot(x = "sepal_length", y="sepal_width", data = phool)
plt.title("Flowers Plot")
plt.show()
```

## Adding Limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris") #iris is a flower dataset
phool

# draw a line plot
sns.lineplot(x = "sepal_length", y="sepal_width", data = phool)
plt.title("Flowers Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Set Styles
- darkgrid

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris") #iris is a flower dataset
phool

# draw a line plot
sns.lineplot(x = "sepal_length", y="sepal_width", data = phool)
plt.title("Flowers Plot")
sns.set_style(style=None, rc=None)
sns.set_style("dark")
plt.show()
```

## Size of Figure


```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris") #iris is a flower dataset

# change figure
plt.figure(figsize=(8,6))
# draw a line plot
sns.lineplot(x = "sepal_length", y="sepal_width", data = phool)
plt.title("Flowers Plot")

plt.show()
```
