# Facecolor follows RGBA Color Format
 - 1st Value represents red color ranging from 0 to 1
 - 2nd Value represents green color ranging from 0 to 1
 - 3rd Value represents blue color ranging from 0 to 1
 - 4th Value represents alpha (opacity) ranging from (transparent) to 1
 - example facecolor = (1,,,1,0,1) is yellow color

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# read a titanic.csv file
# from seaborn library 
kashti1 = sns.load_dataset("titanic") 


sns.barplot(x="class", y="fare", data=kashti1,
           linewidth=3, facecolor=(0.1,0.3,0.3,0.5)
           errcolor=".2", edgecolor=".2") #errcolor 1 is white  edge color bar edge
plt.show()
```
